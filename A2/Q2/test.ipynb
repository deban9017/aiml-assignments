```python
import numpy as np
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score

matrix = np.array([
    [199, 0, 0, 1, 3, 0, 2, 1, 1, 1],
    [0, 192, 1, 0, 1, 2, 1, 0, 5, 0],
    [4, 6, 155, 4, 3, 1, 7, 4, 8, 0],
    [1, 2, 9, 167, 0, 10, 2, 5, 2, 3],
    [0, 2, 1, 0, 191, 3, 4, 0, 2, 6],
    [9, 3, 1, 10, 2, 141, 6, 5, 7, 2],
    [4, 1, 5, 0, 2, 4, 195, 0, 0, 0],
    [0, 7, 3, 2, 3, 1, 0, 180, 2, 6],
    [4, 9, 4, 8, 4, 14, 1, 4, 161, 3],
    [0, 1, 1, 7, 9, 0, 0, 11, 2, 144]
])

true_labels = []
predicted_labels = []
for i in range(matrix.shape[0]):
    for j in range(matrix.shape[1]):
        true_labels.extend([i] * matrix[i, j])
        predicted_labels.extend([j] * matrix[i, j])

for class_label in range(matrix.shape[0]):
    true_class = [1 if label == class_label else 0 for label in true_labels]
    pred_class = [1 if label == class_label else 0 for label in predicted_labels]

    accuracy = accuracy_score(true_class, pred_class)
    precision = precision_score(true_class, pred_class, zero_division=0)
    recall = recall_score(true_class, pred_class, zero_division=0)
    f1 = f1_score(true_class, pred_class, zero_division=0)

    print(f"Class {class_label}:")
    print(f"  Accuracy:  {accuracy:.4f}")
    print(f"  Precision: {precision:.4f}")
    print(f"  Recall:    {recall:.4f}")
    print(f"  F1 Score:  {f1:.4f}")
    print("-" * 15)
```
